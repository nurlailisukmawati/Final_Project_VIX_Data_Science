```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing


import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('regression_dataset.csv')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
cut_off = round(df.shape[0] * 0.9)
df_train = df[:cut_off]
df_test = df[cut_off:].reset_index(drop = True)
df_train.shape, df_test.shape
```

```python
plt.figure(figsize = (20,5))
sns.lineplot(data = df_train, x = df_train['date'], y = df_train['qty'])
sns.lineplot(data = df_test, x = df_test['date'], y = df_test['qty'])
```
